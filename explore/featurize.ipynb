# Household

```python
df1 = get_db('usc_household')
df1 = df1.merge(df2, left_on = 'id2', right_on = 'geoid')
df1 = cut_df(df1)
print df1.shape
df1.head()
```

```python
df1 = df1.groupby(['lat_cut', 'lon_cut']).sum().dropna().reset_index()
print df1.shape
df1.head()
```

```python
total_p = 0
p_range = range(1,8)
for i in p_range:
    col = 'p' + str(i)
    total_p += df1[col] * i
av_p = total_p / df1.total
df1['avg_hh_size'] = av_p
df1.fillna(0, inplace=True)

print df1.shape
df1.head()
```

```python
df1 = df1[['lat_cut', 'lon_cut', 'avg_hh_size']]
print df1.shape
df1.head()
```

```python
df1 = make_feature_df(['usc_household'])
print df1.shape
df1.head()
```
